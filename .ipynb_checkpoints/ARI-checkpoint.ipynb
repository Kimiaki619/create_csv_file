```python
#ARI　クラスタリングの評価方法
from sklearn import metrics
import numpy as np
import pandas as pd

class ARI_kid:
    def __init__(self,csv_label_falie,csv_cla_falie):
        self.label = csv_label_falie
        self.cla = csv_cla_falie
        self.label_data, self.cla_data = self.data()
        
    def data(self):
        df_label = pd.read_csv(self.label,index_col=0)
        df_cla = pd.read_csv(self.cla,index_col=1)
        
        df_label_index = df_label.values
        df_label_index = [df_label_index[f][0] for f in range(len(df_label_index))]
        df_cla = df_cla.reindex(index=df_label_index)
        
        df_label = df_label.values
        df_cla = df_cla.values
        
        df_label = [df_label[f][1] for f in range(len(df_label))]
        df_cla = [df_cla[f][1] for f in range(len(df_cla))]
        
        return df_label,df_cla
        
        
    def ARI(self):
        #A_R_I = metrics.adjusted_rand_score(self.label_data,self.cla_data)
        A_R_I = metrics.normalized_mutual_info_score(self.label_data,self.cla_data)
        #A_R_I = metrics.adjusted_rand_score(self.label_data,self.cla_data)
        return A_R_I

ARI = ARI_kid(csv_label_falie="label_6_train_2.csv",csv_cla_falie="clas_6_gausian_3000_.csv")
print(ARI.label_data)
print(ARI.cla_data)

#for n,i in enumerate(ARI.cla_data):
#    print(n,i)

data = ARI.ARI()
print(data)
```
